```python
# Import all the required libraries and dependencies for the project
import pandas as pd
import hvplot.pandas
from pathlib import Path
import numpy as np
import geoviews as gv
import geopandas as gpd
from bokeh.models import HoverTool
import holoviews as hv
from holoviews import opts
```

```python
#Import the necessary csv data

sfo_data_df = pd.read_csv(
    Path(r"C:\Users\Oluwa\GITHUB\Bootcamp\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources\sfo_neighborhoods_census_data.csv"),)

neighborhood_locations_df=pd.read_csv(
    Path(r"C:\Users\Oluwa\GITHUB\Bootcamp\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources\neighborhoods_coordinates.csv"), index_col='Neighborhood')
```

```python
#Read the CSV Data

print (sfo_data_df.head())
print (sfo_data_df.tail())
```

```python
#Group the dataframe data by the column 'year'
annual_sfo= sfo_data_df.groupby('year')

#Set a new variable for the mean values of each year in each of the columns
yearly_mean= annual_sfo.mean()
```

```python
#Plot a bar chat for the housing units column
yearly_units_bar= yearly_mean.hvplot.bar(x='year', y='housing_units', xlabel='Year', ylabel='Housing Units',
                                   title='San Francisco Housing Units By Year', width=800, height=400)

#Format the Y Axis so it strops using scientific numbers
formatted_bar_chart = yearly_units_bar.opts(yformatter='%.0f')

#Limit the y axis to show better data disparages
sfo_housing_units = formatted_bar_chart.opts(ylim=(365000,385000))

#Show the Bar Graph
sfo_housing_units
```

```python
#Questions about Plot 1

print(f"The overall trend in Housing Units in San Francisco is RISING by an average of about 2,000 units per year")
```

```python
#Create a new dataframe without the housing units column

annual_sqrft= yearly_mean.drop(columns='housing_units')
```

```python
#Plot a line graph that shows the sales price and rent in comparison with one another
yearly_prices_by_sqft= annual_sqrft.hvplot.line(x='year', y=['sale_price_sqr_foot', 'gross_rent'], xlabel='Year', ylabel='Sales Price Per Square Foot / Gross Rent',
                                   title='Sale Price vs Average Rent in SFO', width=800, height=400)

#Show the line graph
yearly_prices_by_sqft
```

```python
#Questions about Plot 2

print(f"2011 saw a slight drop in sales price from the previous year and also had a slower rise in rent increase than many of the years that followed but the rent did increase ")
```

```python
#Group the original dataset by year and neighborhood and make it a new variable

sfo_crossdata= sfo_data_df.groupby(['year', 'neighborhood']).mean().drop(columns='housing_units')
```

```python
#Create a line plot that's filtered by neighborhood 
sfo_plot = sfo_crossdata.hvplot.line(x='year', y=['sale_price_sqr_foot', 'gross_rent'],
                                 xlabel='Year', ylabel='Gross Rent/ Sales Price Per Square Foot', title='SFO by Year by Neighborhood',
                                 line_width=2, width=600, height=400, groupby='neighborhood', legend='top_left')

#Show the line plot
sfo_plot
```

```python
#Questions about Plot 3

print(f"In Anza Vista the sales price in 2016 was lower than 2012 ")
```

```python
#Create a new data frame which groups the original dataset by neighborhoods
sfo_neighborhoods=sfo_data_df.groupby('neighborhood').mean()

#Make sure the columns are named the same so that merge and concat functions can work
neighborhood_locations_df=neighborhood_locations_df.rename_axis('neighborhood')
```

```python
#Concatenate all neighborhood data
all_neighborhoods_df = pd.concat(
    [sfo_neighborhoods, neighborhood_locations_df], 
    axis="columns",
    sort=False
)
#Drop empty columns
all_neighborhoods_df= all_neighborhoods_df.dropna()

#Make sure the years are integers
all_neighborhoods_df['year'] = all_neighborhoods_df['year'].astype(int)

#Print the dataframe
all_neighborhoods_df
```

```python
#Create an OSM Map using the data points

SFO_Map_Plot=all_neighborhoods_df.hvplot.points(
    'Lon', 
    'Lat', 
    geo=True, 
    color='gross_rent',
    size='sale_price_sqr_foot',
    alpha=.6,
    tiles='OSM',
    frame_width = 700,
    frame_height = 500,
    title= 'SFO Neighborhood Affordability Map',
    hover_cols=['neighborhood', 'gross_rent', 'sale_price_sqr_foot'],
    clabel='GROSS RENT'
    )

#Print the Map
SFO_Map_Plot
```

```python
#Questions about Plot 4

#Answer Variables (Map Plot Wasnt Clear In The Required Assignment Format)

High_Rent=all_neighborhoods_df['gross_rent'].idxmax()
High_sale_price=all_neighborhoods_df['sale_price_sqr_foot'].idxmax()

#Answers

print(f"The neighborhood with the highest gross rent:", {High_Rent})
print(f"The neighborhood with the highest sale price:", {High_sale_price})
```

```python
#I tried to reorganize the hover information but it didnt work

def custom_hover_settings():
    hover = HoverTool()
    hover.tooltips = [
        ("Neighborhood", "@neighborhood"),
        ("Gross Rent", "@gross_rent{0,0.00}"),
        ("Sale Price per Sq. Foot", "@sale_price_sqr_foot{0,0.00}"),
    ]
    return hover

Cleaned_SFO_Map = SFO_Map_Plot.opts(show_legend=True, tools=[custom_hover_settings])



#Define the current minimum and maximum 'housing units' value and the minimum and maximum I would like to use for the map
min_value = all_neighborhoods_df['housing_units'].min()
max_value = all_neighborhoods_df['housing_units'].max()
new_min = .1
new_max = .5

#Normalize the housing units data so the datapoints arent massive
all_neighborhoods_df['normalized_housing_units'] = (
    (new_min + (new_max - new_min) * (all_neighborhoods_df['housing_units']))*min_value/175930000000)
                                     
                                     
#Create a new map where the size is determined by number of housing units
SFO_Map_Plot2=all_neighborhoods_df.hvplot.points(
    'Lon', 
    'Lat', 
    geo=True, 
    color='gross_rent',
    size='sale_price_sqr_foot',
    alpha='normalized_housing_units',
    tiles='OSM',
    frame_width = 700,
    frame_height = 500,
    title= 'SFO Neighborhood Affordability Map',
    hover_cols=['neighborhood', 'gross_rent', 'sale_price_sqr_foot'],
    clabel='GROSS RENT',
    )

#Print the Map
SFO_Map_Plot2
```
